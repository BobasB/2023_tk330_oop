- конкатенація (додавання/зчепленя) та множення стрічок

```python
c = a + "| Це буде рядок зчелення |" + b
d = a.__add__("| Це буде рядок зчелення |".__add__(b))

print("В обох випадках будемо мати:\n", c, "\n", d)

# Ці два записи рівноціні
e = a * 4
f = a.__mul__(4)
print("Що буде при множенні стрічки на число?\n", e, "\n", f)
# Найпростіше можна побавитись у вивід стрічки символів заповнювачів
print("~=+=~" * 20)
print(" " * 5 + "!" * 1 + " " * 5)
```

- Запитались в ChatGPT як вивести піраміде з символів
>Для створення піраміди з символів "!" можна використовувати цикли. Ось приклад коду на Python, який буде виводити піраміду, починаючи з вашого початкового рядка:

```python
spaces = 5  # Початкова кількість пробілів
exclamation_marks = 1  # Початкова кількість знаків оклику

# Внешній цикл для кожного рядка піраміди
for i in range(5):  # Змініть 5 на бажану висоту піраміди
    # Друкуємо пробіли
    print(" " * spaces, end="")
    
    # Друкуємо знаки оклику
    print("!" * exclamation_marks, end="")
    
    # Збільшуємо кількість знаків оклику і зменшуємо кількість пробілів для наступного рядка
    exclamation_marks += 2
    spaces -= 1
    
    # Перехід на новий рядок
    print()
```

- числові типи даних

```python
a = 5
b = int(5)

print("Цілі числа автоматично будуть відноситись до класу:", type(a), type(b))
c = 5.5
d = float(a)

print("Якщо число є дробовим то буде тип:", type(c), 
      "\nАбо можна явно перетворити", type(a), "в", type(d),
      "\nПісля перетворення число", a, "тепер буде", d)
```

```python
b += 0.1 # Це ще називають накопичуванням
print("В цьому випадку маємо неявне перетворення: ", type(b), "\nРезультат (буде з похибкою):", b)

e = a + 0.1
f = 0.1 + a
print("\nТут все нормально працює:", type(e), "  значення:", e,
      "\nТут також результат буде: ", type(f), "  значення:", f,
      )

e = a.__add__(0.1)
f = 0.1.__add__(a)
print("З використанням методів класу додавання буде працювати тільки за певних умов.",
      "\nДля цілих типів даних можна додавати тільки цілі, інакше результат буде:", type(e), "  значення:", e,
      "\nЯкщо ж цілі числа будем додавати до з плаваючою крапкою то буде: ", type(f), "  значення:", f,
      )
```

- знайомимось з списками та кортеджі
> `List` - впорядкована колекція змінних типів даних; 

> `Tuple` - впорядкована колекція незмінних даних;

```python
l = [1, 1.2, "текст", a]
ll = list(l)

print("Початкове значення списку:",l)

l.append("Це кінець") # Всталяємо в кінець
l.insert(1, "Це буде на індексі 1") # Вставляємо на позицію 1
l.pop(-2) # видаляємо 5

print("Після змін, список має вигляд:",l)

l.extend(ll)
print("Якщо зчепити два списки, буде:",l)
```

```python
t = (1, 1.2, "текст", a, 1)
tt = tuple(t)

print("Початкове значення кортеджу:",t)

print("Навіть використовуючи зчеплення, ми не модифікували наш tuple, а просто повернули нове значення:", t.__add__((1,)))

print("Кількість 1 буде:",t.count(1))
print("Знайдемо під яким індексом знаходиться число 1:",t.index(1))
```
