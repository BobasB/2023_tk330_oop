- Запитались в ChatGPT як вивести піраміде з символів
>Для створення піраміди з символів "!" можна використовувати цикли. Ось приклад коду на Python, який буде виводити піраміду, починаючи з вашого початкового рядка:

```python
spaces = 5  # Початкова кількість пробілів
exclamation_marks = 1  # Початкова кількість знаків оклику

# Внешній цикл для кожного рядка піраміди
for i in range(5):  # Змініть 5 на бажану висоту піраміди
    # Друкуємо пробіли
    print(" " * spaces, end="")
    
    # Друкуємо знаки оклику
    print("!" * exclamation_marks, end="")
    
    # Збільшуємо кількість знаків оклику і зменшуємо кількість пробілів для наступного рядка
    exclamation_marks += 2
    spaces -= 1
    
    # Перехід на новий рядок
    print()
```

```python
l = [1, 1.2, "текст", a]
ll = list(l)

print("Початкове значення списку:",l)

l.append("Це кінець") # Всталяємо в кінець
l.insert(1, "Це буде на індексі 1") # Вставляємо на позицію 1
l.pop(-2) # видаляємо 5

print("Після змін, список має вигляд:",l)

l.extend(ll)
print("Якщо зчепити два списки, буде:",l)
```

- робота з циклами
> `while` - умовний цикл - ітерації будуть виконуватись поки умова є істииною (може бути безмежним циклом)

> `for` - із задою кількістю ітерацій (повторів)

```python
from random import randint
n = ["Богдан", "Сергій", "Василь"]
h = ["футбол","аніме","програмування"]

j = 0
while j < 3:
    print(">>> {name} любить {hobbi}".format(name=n[j],hobbi=h[randint(0,2)]))
    j+=1

print("="*10, "Це вже не входитиь у цикл, цикл закінчився.\nЦикл While не дуже підходить для визначеної кількості ітерацій.")

# навіть для for можна йти по різному
for j in range(len(n)):
    print(">>> {name} любить {hobbi}".format(name=n[j],hobbi=h[randint(0,2)]))

print("="*10,"Це вже виконається поза циклом")

print(f"Демонструємо що буде виконувати функція range: {list(range(3))}")

# варіант 2 для for (це вже правильне використання for)
for j in n:
    print(">>> {name} любить {hobbi}".format(name=j,hobbi=h[randint(0,2)]))

# Спростимо код ще більше
print("="*10)
for j in n:
    print(f">>> {j} любить {h[randint(0,2)]}")
```

> пробуємо використати цикли при симуляції кидання кубика

```python
# числа можуть бити тільки від 1 до 5, і ми їх можемо вводити з клавіатури
required_number = 5 #int(input("Введіть бажане значення числа на кубику (від 1 до 6):"))

dice_number = int()
print("Значення яка повинно випасти на кубику відповідає умові від 1 до 6?")
if 1 <= required_number <= 6:
    print("Введено коректні дані, починаємо кидати кубик:")
    while required_number != dice_number: # задаємо умову неістинності
        dice_number = randint(1,6)
        print(f"Кидаємо кубик, випало значення {dice_number}. Отримане значення по відношенню до бажаного буде: {required_number == dice_number}.")
else:
    print(f"Введено не коректне значення: {required_number}")
```

```python
attempts = 4 # Кількіть спроб
required_number = 6 # число яке має випасти на кубику

dice_number = int()

# якщо ми не будео використовувати нашу ітераційну змінну у циклі, то можемо її назвати _ (просто нижнє підкреслення)  
for _ in range(attempts):
    dice_number = randint(1,6)
    print(f"Кидаємо кубик, випало значення {dice_number}. Отримане значення по відношенню до бажаного буде: {required_number == dice_number}.")
    if required_number == dice_number:
        print(f"Умова виконана на {_ + 1} ітерації, виходимо з циклу.")
        break
else:
    print("Задане значення не було отримано. Ви програли.")
```
