## Загальні поняття
> клас - це загальний темплейт який описує щось конкретне;

> обєкт - це окремий екземпляр класу з своїми характеристиками/параметрами;

> тип даних - характеристика набору даних;

- вбудовані типи даних (текстові/стрічкові)

```python
print("Це тектові дані які будуть виведені")
a = "Ця змінна мість текстові дані"
print(a)
b = str("Це будуть текстові дані через клас str")
print(b)

print("Рахуємо скільки раз зустрічається буква е в обєкті b: ", b.count("е"))
print("Рахуємо скільки раз зустрічається буква і в текстовій змінній a: ", a.count("і"))
print("Рахуємо скільки раз зустрічається буква і в простій стрічці: ", "Це буде стрічка без присвоєння змінній чи створення обєкту".count("і"))
```
